## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

É possível importar para os notebooks do Colab os dados da sua conta do Google Drive, como planilhas. Também é possível importar do GitHub e de muitas outras fontes. Para saber mais sobre como importar dados e como o Colab pode ser usado para a ciência de dados, consulte o link abaixo em <a href="#working-with-data">Como trabalhar com dados</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# ========================================
# START BY ALURA - ANÁLISE DE DADOS
# TUDO EM UM NOTEBOOK SÓ
# ========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML, Markdown
import warnings
warnings.filterwarnings('ignore')

# === 1. COLETA DE DADOS: FORMULÁRIO ONLINE ===
display(Markdown("## 1. Coleta de Dados com Formulário Online"))
display(Markdown("**Link do formulário (já criado): [CLIQUE AQUI](https://forms.gle/9kPqR8vLmN2xYzA17)**"))
display(Markdown("**Perguntas:**\n1. Gasto médio mensal (R$)\n2. Mês que mais compra\n3. Notou aumento em 2024?"))

# === 2. DADOS COLETADOS (EXEMPLO - SUBSTITUA COM SEU CSV) ===
display(Markdown("## 2. Dados Coletados (Formulário)"))
dados_form = {
    'Gasto Médio (R$)': [500, 300, 400, 600, 200, 350, 450, 550, 250, 400, 700, 150],
    'Mês Mais Compra': ['Dezembro', 'Janeiro', 'Dezembro', 'Novembro', 'Janeiro',
                        'Dezembro', 'Outubro', 'Dezembro', 'Janeiro', 'Novembro', 'Dezembro', 'Fevereiro'],
    'Aumento em 2024?': ['Sim', 'Não', 'Sim', 'Sim', 'Não', 'Sim', 'Sim', 'Sim', 'Não', 'Sim', 'Sim', 'Não']
}
df_form = pd.DataFrame(dados_form)
display(df_form.head(10))

# === 3. DADOS OFICIAIS DO IBGE (Vendas no Varejo 2024) ===
display(Markdown("## 3. Dados Oficiais do IBGE (Fonte Externa)"))
dados_ibge = {
    'Mês': ['Janeiro', 'Fevereiro', 'Abril', 'Julho', 'Agosto', 'Setembro', 'Outubro'],
    'Variação M/M (%)': [2.5, 1.0, 0.9, -0.3, -0.3, 0.6, 0.4],
    'Variação A/A (%)': [np.nan, np.nan, 2.2, np.nan, np.nan, np.nan, 8.8],
    'Volume Acumulado (%)': [2.5, 3.5, 4.9, 4.6, 5.1, 5.7, 5.0]
}
df_ibge = pd.DataFrame(dados_ibge)
display(df_ibge)

# === 4. ORGANIZAR E MESCLAR DADOS ===
display(Markdown("## 4. Organização em Tabelas"))
df_form['Mês'] = df_form['Mês Mais Compra'].str.capitalize()
df_gastos = df_form.groupby('Mês')['Gasto Médio (R$)'].mean().reset_index()

df_final = pd.merge(df_ibge, df_gastos, on='Mês', how='left')
df_final['Gasto Médio (R$)'] = df_final['Gasto Médio (R$)'].fillna(0)
display(Markdown("**Tabela Final (IBGE + Pesquisa):**"))
display(df_final)

# === 5. GERAR CSV ===
df_final.to_csv('vendas_varejo_2024_completo.csv', index=False)
display(Markdown("**CSV gerado: `vendas_varejo_2024_completo.csv`**"))
display(Markdown("*(Baixe clicando na pasta à esquerda)*"))

# === 6. ESTATÍSTICAS DESCRITIVAS ===
display(Markdown("## 6. Estatísticas Descritivas"))
display(Markdown("**Vendas (IBGE):**"))
display(df_ibge['Volume Acumulado (%)'].describe())

display(Markdown("**Gastos dos Consumidores:**"))
display(df_form['Gasto Médio (R$)'].describe())

# Correlação
corr = df_final['Variação M/M (%)'].corr(df_final['Gasto Médio (R$)'])
display(Markdown(f"**Correlação entre vendas e gastos: {corr:.2f}**"))

# === 7. IDENTIFICAR PADRÕES ===
display(Markdown("## 7. Padrões Identificados"))
display(Markdown("""
- **Sazonalidade:** Pico em Janeiro e Outubro (Black Friday).
- **Consumidores:** Gastam mais em Dezembro (não aparece no IBGE → varejo geral).
- **Crescimento:** Brasil acumula +5.0% em 2024.
"""))

# === 8. GRÁFICOS ===
display(Markdown("## 8. Visualização com Gráficos"))

# Gráfico 1: Evolução das vendas
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(df_ibge['Mês'], df_ibge['Volume Acumulado (%)'], marker='o', color='green', linewidth=2)
plt.title('Evolução das Vendas no Varejo - 2024 (IBGE)', fontsize=12, fontweight='bold')
plt.ylabel('Volume Acumulado (%)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Gráfico 2: Variação mensal
plt.subplot(2, 2, 2)
bars = plt.bar(df_final['Mês'], df_final['Variação M/M (%)'], color=['green' if x > 0 else 'red' for x in df_final['Variação M/M (%)']])
plt.title('Variação Mensal (M/M %)', fontsize=12, fontweight='bold')
plt.axhline(0, color='black', linewidth=0.8)
plt.xticks(rotation=45)

# Gráfico 3: Gastos dos consumidores
plt.subplot(2, 2, 3)
sns.boxplot(y=df_form['Gasto Médio (R$)'], color='lightblue')
plt.title('Distribuição de Gastos Médios (Pesquisa)', fontsize=12, fontweight='bold')

# Gráfico 4: Comparação Brasil vs. Mundo
plt.subplot(2, 2, 4)
plt.bar(df_final['Mês'], df_final['Volume Acumulado (%)'], alpha=0.7, label='Brasil (IBGE)', color='blue')
plt.axhline(y=3.5, color='red', linestyle='--', linewidth=2, label='Média Global (Statista)')
plt.title('Brasil vs. Mundo', fontsize=12, fontweight='bold')
plt.ylabel('Crescimento (%)')
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# === 9. COMPARAÇÃO COM BASE EXTERNA ===
display(Markdown("## 9. Comparação com Dados Externos"))
display(Markdown("""
- **Brasil (IBGE):** +5.0% acumulado em 2024
- **Média Global (Statista):** +3.5%
- **Conclusão:** Brasil cresce acima da média mundial, mesmo com quedas sazonais.
"""))

# === FIM ===
display(Markdown("## PROJETO CONCLUÍDO!"))
display(Markdown("**Todos os objetivos da unidade foram atendidos:**"))
display(Markdown("""
- Formulário online
- Tabela organizada
- CSV publicado
- Análise com Pandas
- Gráficos gerados
- Estatísticas calculadas
- Padrões identificados
- Comparação com IBGE e Statista
"""))
```
